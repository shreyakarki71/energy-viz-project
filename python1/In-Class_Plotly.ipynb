```python
import pandas as pd

energy = pd.read_csv('https://raw.githubusercontent.com/owid/energy-data/refs/heads/master/owid-energy-data.csv')

energy
```

```python
import plotly.express as px

energy = energy[energy['year'] == 2000].dropna(subset=["solar_share_energy", "gdp"])

# create solar ratio column: GDP divided by solar energy usage
energy['solar ratio'] = energy['gdp'] / energy['solar_share_energy']

energy
```

```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
# basic cleaning 
df = energy.copy()
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.replace([np.inf, -np.inf], np.nan)

df.loc[df["gdp"] == 0, "gdp"] = np.nan
df.loc[df["population"] == 0, "population"] = np.nan

df["gdp_per_capita"] = df["gdp"] / df["population"]

latest_year = df["year"].max()
df_latest = df[df["year"] == latest_year]
```

**GDP per capita vs renewables share**


```python
fig1 = px.scatter(
    df_latest.dropna(subset=["gdp_per_capita", "renewables_share_energy", "population"]),
    x="gdp_per_capita",
    y="renewables_share_energy",
    size="population",
    hover_name="country",
    log_x=True,
    size_max=35,
    opacity=0.8,
    title=f"GDP per Capita vs Renewable Share of Energy ({latest_year})"
)

# hovertemplate to show country name, GDP per capita, and renewable share with formatting
fig1.update_traces(
    marker=dict(line=dict(width=0.5, color="white")),
    hovertemplate="<b>%{hovertext}</b><br>"
                  "GDP per Capita: %{x:,.0f}<br>"
                  "Renewables Share: %{y:.2f}%<br>"
                  "<extra></extra>"
)

# set x-axis to log scale with custom ticks and labels
fig1.update_xaxes(
    type="log",
    tickvals=[1000, 5000, 10000, 20000, 50000],
    ticktext=["1k", "5k", "10k", "20k", "50k"]
)

fig1.update_layout(
    template="plotly_white",
    xaxis_title="GDP per Capita (USD, log scale)",
    yaxis_title="Renewable Share of Energy (%)",
    margin=dict(t=80),
    height=600
)

fig1.show()
```

**Countries relying on imported electricity**

```python
#  latest year in dataset
latest_year = int(df["year"].max())

# filter to latest year and drop rows with missing iso_code or net_elec_imports_share_demand
map_df = df[df["year"] == latest_year].dropna(
    subset=["iso_code", "net_elec_imports_share_demand"]
).copy()

# Remove aggregate entities (OWID codes) that don't correspond to individual countries
map_df = map_df[~map_df["iso_code"].astype(str).str.contains("OWID", na=False)]

# Clip net imports to -20% to 20% for better color scaling
map_df["imports_clipped"] = map_df["net_elec_imports_share_demand"].clip(-20, 20)

# Create choropleth map
fig = px.choropleth(
    map_df,
    locations="iso_code",
    color="imports_clipped",
    hover_name="country",
    color_continuous_scale=["#8B0000", "#FF4C4C", "#FFFFFF", "#4C8BFF", "#0033CC"],
    color_continuous_midpoint=0,
    range_color=(-20, 20),
    title=f"Net Electricity Imports as % of Demand ({latest_year})"
)

# Make map fill space nicely
fig.update_layout(
    template="plotly_white",
    height=600,
    margin=dict(l=0, r=0, t=70, b=60)
)

fig.update_geos(
    fitbounds="locations",
    visible=False
)

# Move color bar to bottom
fig.update_coloraxes(
    colorbar=dict(
        orientation="h",
        y=-0.2,
        x=0.5,
        xanchor="center",
        title="Net imports (% demand)"
    )
)

fig.show()
```

```python
latest_year = int(df["year"].max())

countries = ["Brazil", "China", "Germany", "India", "United States"]

mix_df = df[(df["year"] == latest_year) & (df["country"].isin(countries))].copy()

mix_df = mix_df[[
    "country",
    "hydro_share_energy",
    "wind_share_energy",
    "solar_share_energy",
    "biofuel_share_energy"
]].dropna()

# Rename columns 
mix_df = mix_df.rename(columns={
    "hydro_share_energy": "Hydro",
    "wind_share_energy": "Wind",
    "solar_share_energy": "Solar",
    "biofuel_share_energy": "Biofuel"
})

# Convert to long format
mix_long = mix_df.melt(
    id_vars="country",
    var_name="Source",
    value_name="Share (%)"
)

fig3 = px.bar(
    mix_long,
    x="country",
    y="Share (%)",
    color="Source",
    title=f"3) Renewable Energy Mix by Source ({latest_year})"
)

fig3.update_layout(
    template="plotly_white",
    xaxis_title="Country",
    yaxis_title="Share of Total Energy (%)",
    legend_title="Renewable Source",
    height=600
)

fig3.update_xaxes(tickangle=-20)

fig3.show()
```
