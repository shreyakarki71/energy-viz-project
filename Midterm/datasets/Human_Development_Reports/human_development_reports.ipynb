```python
# Human Development Reports
df = pd.read_csv('./HDR25_Composite_indices_complete_time_series.csv', encoding='latin-1')

df
```

```python
meta = pd.read_excel('./HDR25_Composite_indices_metadata.xlsx', sheet_name='codebook')

meta
```

```python
pattern = r'\d{4}-\d{4}'  # Matches any 4-digit year
mask = meta['Time series'].str.contains(pattern, regex=True, na=False)
# Get the field names which have time series data for a range of years
fields = meta.loc[mask, 'Full name'].tolist()

fields
```

```python
import re

# Rename columns based on the metadata, replacing short names with long names but keeping the year so we can melt
df_renamed = df.rename(columns={col: re.sub(r'(.*)_(\d{4})', lambda m: meta.loc[meta['Short name'] == m.group(1), 'Full name'].values[0] + f": {m.group(2)}", col) for col in df.columns})
df_renamed
```

```python
# manipulate the dataframe to get the fields melted into long format (years in a single column, values in other column)
df_long = pd.wide_to_long(df_renamed, stubnames=fields,
                          i=['iso3'], 
                          j='Year', 
                          sep=': ').reset_index()

df_long
```

```python
# Now, make a separate aggregates dataframe

aggregates = df_long[df_long['iso3'].str.contains('\\.')].reset_index(drop=True)
aggregates.dropna(axis=1, how='all', inplace=True)
aggregates
```

```python
# Do the reverse for the country-level data.

df_long = df_long[~df_long['iso3'].str.contains('\\.')].reset_index(drop=True)
df_long
```

```python
# output the data for use in Tableau if you want

df_long.to_csv('./human_development.csv', index=False)
aggregates.to_csv('./human_development_aggregates.csv', index=False)
```
