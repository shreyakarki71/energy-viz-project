```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 1, 3, 5]})
fig = px.line(df, x='x', y='y', title='Simple Plotly Demo')
fig.show()
```
